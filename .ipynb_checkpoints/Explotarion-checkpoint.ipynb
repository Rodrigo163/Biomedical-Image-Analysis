## N-dimensional images

```python
#creating a volume image from 3 diff images
import numpy as np
im1 = iio.imread('tcia-chest-ct-sample/chest-220.dcm')
im2 = iio.imread('tcia-chest-ct-sample/chest-221.dcm')
im3 = iio.imread('tcia-chest-ct-sample/chest-222.dcm')

vol = np.stack([im1, im2, im3])
vol.shape
```

```python
#if we already have the 2d images and want to create a volume
#to check the files
import os
os.listdir('tcia-chest-ct-sample')
```

```python
vol = iio.volread('tcia-chest-ct-sample')
#this function will check if there is an order in the metadata
#if not, it will stack them alphabet. 
```

```python
#image shape
n0,n1,n2
```

```python
#sampling rate (in mm) = amount of space covered by each element
d0, d1, d2 = vol.meta['sampling']
d0, d1, d2
```

```python
#field of view (in mm) = total amount of space covered along each axis
n0*d0, n1*d1, n2*d2
```

## Advanced Plotting
To plot N-dimensional data slice it!

```python
#plotting multiple images at once with plt.subplots
fig, axes = plt.subplots(nrows=1, ncols=3)
i = 0
for ax in axes:
    axes[i].imshow(vol[i], cmap='gray')
    ax.axis('off')
    i += 1
```

Depending on the direction we slice the data on, the views have different names. 
- axial
- coronal
- sagittal

Sometimes images don't have same rate along each axes so we'll have to strech them to account for this. 

```python
#we can get the sampling rate from the metadata
# store the ratios in some variable and then to adjust distortion 
# plt.imshow(im, cmap= 'gray', aspect = asp)
```

To select a 2D frame, pick a frame for the first axis and select all data from the remaining two: vol[0, :, :]

For this exercise, use for loop to plot every 40th slice of vol on a separate subplot. matplotlib.pyplot (as plt) has been imported for you.

```python
# Plot the images on a subplots array 
fig, axes = plt.subplots(nrows=1, ncols=4)

# Loop through subplots and draw image
#vol[5] = vol[5,:,:]
for ii in range(4):
    im = vol[ii*40]
    axes[ii].imshow(im, cmap='gray')
    axes[ii].axis('off')
    
# Render the figure
plt.show()
```
