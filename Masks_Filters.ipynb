### Equalization
distributions often skewed toward low intensities (because of background values). 
- Equalization: redistribute values to optimize full intensity range
- CDF (cumulative distribution function): shows proportion of pixels in range
There is always a trade-off! Be careful not to take away too much info by doing this.

```python
cdf = hist.cumsum() / hist.sum()
cdf.shape
```

```python
im_equalized = cdf[im] * 255
fig, axes = plt.subplots(2,1)
axes[0].imshow(im)
axes[1].imshow(im_equalized)
plt.show()
```

## Masks
To restrict analysis to important parts of image.

A mask is a boolean array that serves as a screen to remote undesirable pixels. They have to retain same dimension of original array so they can be overlaid. 

In the histogram we see a step drop after ~40. Let's select values bigger than this. 

```python
mask1 = im>42
plt.imshow(mask1)
```

```python
#bone is the highest intensity tissue in X-ray so if we increase the filter 
#we'll create a rough bone mask
mask2 = im>64
plt.imshow(mask2)
```

```python
#mask of none bone tissue
mask3 = mask1 & ~mask2 
plt.imshow(mask3)
```

```python
#using np.where to create nasks
#np.where(condition, x, y) will return x when true and y when false
im_bone = np.where(im > 64, im, 0)
plt.imshow(im_bone, cmap='gray')
plt.axis('off')
plt.show()
```

# Tune a mask
Imperfect masks can be tuned through the addition and subtraction of pixels. SciPy includes several useful methods for accomplishing these ends. These include:

- binary_dilation: Add pixels along edges
- binary_erosion: Remove pixels along edges
- binary_opening: Erode then dilate, "opening" areas near edges
- binary_closing: Dilate then erode, "filling in" holes
For this exercise, create a bone mask then tune it to include additional pixels.

```python
# Create and tune bone mask
mask_bone = im >=145
mask_dilate = ndi.binary_dilation(mask_bone, iterations=5)
mask_closed = ndi.binary_closing(mask_bone, iterations=5)

# Plot masked images
fig, axes = plt.subplots(1,3)
axes[0].imshow(mask_bone)
axes[1].imshow(mask_dilate)
axes[2].imshow(mask_closed)
#format_and_render_plot()
plt.show()
```

## Filters
Combining spacial and intensity info by using convolutional filters. Examples: 
- smoothing: emphasis large intensity paterns by reducing variability bet. neighbouring pixels. Supresses noise by blurring the image

- sharpening: sharp changes are enhanced, exagerating diff. bet. pixels. 
We can also create our own convolutional filters with ndi

```python
im = iio.imread('hand.png')
#with this we're gonna average pixel with it's neighbours
#this is the kernel
weights = [[.11, .11, .11], 
           [.11, .12, .11], 
           [.11, .11, .11]]
im_filt = ndi.convolve(im, weights)

fig, axes = plt.subplots(2,1)
axes[0].imshow(im, cmap='gray')
axes[1].imshow(im_filt, cmap='gray')
plt.show()
```

We can also use functions as kernels. scipy.ndimage.filters includes:
- median_filter()
- uniform_filter()
- maximum_filter()
- percentile_filter()
for any filter or kernel we can also decide how many neighbours it'll take into acc

```python
im_med = ndi.median_filter(im, size=10)
plt.imshow(im_med)
```

To smooth data across large areas we can use a Gaussian filter --> filter weights dissipate in a circular pattern as we get away from the center. We control the width of the distrib with sigma. Be careful with sigma!

```python
im_gau=ndi.gaussian_filter(im, sigma=4)
plt.imshow(im_gau)
```

## Feature Detection
We can also use filters to detect zones of interest like edges.

Original * Kernel = Edges

An edge is a change in intensity along an axis.
Sharp edge = very high contrast. Our filter is gonna have to reflect this.


```python
#Edge detection
im = iio.imread('hand.png')
im = np.uint8(im)
#we construct a kernel that will look for changes in intensity from top to bottom
#This filter will calculate the diff bet top and bottom rows
weights = [[1, 1, 1],
           [0, 0, 0], 
           [-1, -1, -1]]
edges = ndi.convolve(im, weights)
plt.imshow(edges, cmap= 'seismic')
```

Note how this filter only detects changes from top to bottom (and negatively from bottom to top) but won't recognize horizontal changes.

Another type are Sobel filters (H and V). They provide more weight to the center pixel.

```python
im_sobel1= ndi.sobel(im, axis=0)
im_sobel2= ndi.sobel(im, axis=1)
fig, axes = plt.subplots(1,2)
axes[0].imshow(im_sobel1, cmap='seismic')
axes[1].imshow(im_sobel2, cmap='seismic')
plt.show()
```
