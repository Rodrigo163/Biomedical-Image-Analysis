## Image intensities
Masks and filters.

In order to use them well we need to understand our data's distribution of intensities

Pixels and voxels (3D volume elements) have two properties: intensity and location. Depending the type of image the meaning of intensity changes (brighter in X-ray means dense stuff that better absorbs radiation).

The range of these values depend on the data type:
- uint 8 --> 256 values from 0-255
- int 8 --> 256 values from -128 to 127
- uint16 --> 2^16 values from 0 to...
with im.size we can see how much memory each type occupies. 

We'll be working with histograms counting number of pixels at each intensity value

```python
im.dtype
```

```python
#we'll change it to uint8 to make it more practical
fig, axes = plt.subplots(nrows= 1, ncols = 2)
axes[0].imshow(im, cmap='gray')
plt.axis('off')
im = np.uint8(im)
axes[1].imshow(im, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import scipy.ndimage as ndi
#if we were working with uint8
hist = ndi.histogram(im, min = 0, max = 255, bins=256)
hist.shape
```

```python
plt.plot(hist)
plt.show()
```

### Equalization
distributions often skewed toward low intensities (because of background values). 
- Equalization: redistribute values to optimize full intensity range
- CDF (cumulative distribution function): shows proportion of pixels in range
There is always a trade-off! Be careful not to take away too much info by doing this.

```python
cdf = hist.cumsum() / hist.sum()
cdf.shape
```

```python
im_equalized = cdf[im] * 255
fig, axes = plt.subplots(2,1)
axes[0].imshow(im)
axes[1].imshow(im_equalized)
plt.show()
```

### Intensity
In this chapter, we will work with a hand radiograph from a 2017 Radiological Society of North America competition. X-ray absorption is highest in dense tissue such as bone, so the resulting intensities should be high. Consequently, images like this can be used to predict "bone age" in children.

To start, let's load the image and check its intensity range.

The image datatype determines the range of possible intensities: e.g., 8-bit unsigned integers (uint8) can take values in the range of 0 to 255. A colorbar can be helpful for connecting these values to the visualized image.

```python
# Load the hand radiograph
im = iio.imread('hand.png')
print('Data type:', im.dtype)
print('Min. value:', im.min)
print('Max value:', im.max)

# Plot the grayscale image
plt.imshow(im, vmin = 0, vmax = 255)
plt.colorbar()
#they used this costum function
#probably to simply take axes off and display in gray
#format_and_render_plot() 
```

```python
#to display hist and cdf together
# Import SciPy's "ndimage" module
import scipy.ndimage as ndi 

# Create a histogram, binned at each possible value
hist = ndi.histogram(im, min = 0, max = 255, bins = 256)

# Create a cumulative distribution function
cdf = hist.cumsum() / hist.sum()

# Plot the histogram and CDF
fig, axes = plt.subplots(2, 1, sharex=True)
axes[0].plot(hist, label='Histogram')
axes[1].plot(cdf, label='CDF')
plt.show()
#format_and_render_plot()
```

## Masks
To restrict analysis to important parts of image.

A mask is a boolean array that serves as a screen to remote undesirable pixels. They have to retain same dimension of original array so they can be overlaid. 

In the histogram we see a step drop after ~40. Let's select values bigger than this. 

```python
mask1 = im>42
plt.imshow(mask1)
```

```python
#bone is the highest intensity tissue in X-ray so if we increase the filter 
#we'll create a rough bone mask
mask2 = im>64
plt.imshow(mask2)
```

```python
#mask of none bone tissue
mask3 = mask1 & ~mask2 
plt.imshow(mask3)
```

```python
#using np.where to create nasks
#np.where(condition, x, y) will return x when true and y when false
im_bone = np.where(im > 64, im, 0)
plt.imshow(im_bone, cmap='gray')
plt.axis('off')
plt.show()
```

Since data is noisy we'll have to tune our masks.
To increase size of mask we can add pixels at the edges. This is called dilation. This is very useful for when edges are fuzzy or to make sure we don't mask out pixels we care about.


```python
ndi.binary_dilation(im, iterations=5)
#this converts all adjacent pixels to the mask into mask pixels
```

The opposite operation is called binary erosion and we use it to cut down the mask into its more central pixels

```python
ndi.binary_erosion(im, iterations=5)
```
